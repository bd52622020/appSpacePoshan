```python
Itemcolumns = ["MovieId", "MovieTitle", "ReleaseDate", "VideoReleaseDate",
                  "IMDbURL", "Unknown", "Action", "Adventure", "Animation",
                  "Children", "Comedy", "Crime", "Documentary", "Drama", "Fantasy",
                  "Film-Noir", "Horror", "Musical", "Mystery", "Romance", "Sci-Fi",
                  "Thriller", "War", "Western"]
df_Items = pd.read_csv("u.item", encoding = "ISO-8859-1", delimiter = "|", names = Itemcolumns )
df_Items
```

## Print a list of the 10 movies that received the most number of ratings

```python
df_ratingData.columns =["UserId", "MovieId", "Rating", "TimeStamp"]
df_itemWithRating = pd.merge(df_ratingData,df_Items, how = "inner")
df_itemWithRating = df_itemWithRating[["MovieTitle","Rating"]].groupby("MovieTitle").sum().nlargest(10, ['Rating'])
df_itemWithRating.sort_index()
```

# 10 movies that received the most number of ratings, sorted by the number of ratings.


```python
df_itemWithRating
```

## list of the number of ratings received by each genre

```python
df_itemWithRating = pd.merge(df_ratingData,df_Items, how = "inner")
dfRatingByGener = df_itemWithRating.loc[:,"Unknown":"Western"].sum().transpose()
pd.DataFrame(dfRatingByGener, columns = ['Rating']).sort_values("Rating", ascending = False)
```

# Print the oldest movie with a “5” rating. 

```python
from datetime import datetime
df_itemWithRating = pd.merge(df_ratingData,df_Items, how = "inner")
df_itemWithRating['ReleaseDate'] = pd.to_datetime(df_itemWithRating['ReleaseDate'])
DfWithRating5 = df_itemWithRating.loc[df_itemWithRating['Rating'] == 5].sort_values("ReleaseDate", ascending = True)
pd.DataFrame(DfWithRating5, columns = ["MovieId", "Rating", "MovieTitle", "ReleaseDate","IMDbURL"]).iloc[:1,:]
```

## A list of the genre of the top 10 most rated movies

```python
aa = df_itemWithRating["MovieId"].value_counts().index[:10]
for i in aa:
    movieTitle = df_itemWithRating.loc[df_itemWithRating["MovieId"] == i,"MovieTitle"].values[0]
    movieSelectedGener = df_itemWithRating[df_itemWithRating["MovieId"] == i].loc[:,"Unknown":"Western"]
    generName = movieSelectedGener.columns[(movieSelectedGener == 1).values[0]]
    print(movieTitle + " : " + ",".join(generName))
    
```
